```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import numpy as np
import pandas as pd
import time
from selenium.common.exceptions import TimeoutException
```

```python
data = pd.read_csv('nasdaq_stocks_data_2.csv')
stocks = list(data['Symbol'])
driver = webdriver.Chrome()
```

```python
base_url = 'https://www.cnbc.com/quotes/{}'
news1, news2, news3, news4, news5 = [], [], [], [], []

stocks = [str(stock) for stock in stocks]

for stock in stocks:
    print('------- ' + stock + ' -------')
    url = base_url.format(stock)
    driver.get(url)
    time.sleep(2)
    all_news = driver.find_elements(By.CLASS_NAME, 'LatestNews-item')[0:5]
    
    # Try-except for each news item and append 'NaN' as a string if not found
    try:
        first_new_text = all_news[0].find_element(By.CLASS_NAME, 'LatestNews-headline').text
        news1.append(first_new_text)
    except Exception as e:
        print(f"Error extracting first news for {stock}: {e}")
        news1.append('NaN')
        
    try:
        second_new_text = all_news[1].find_element(By.CLASS_NAME, 'LatestNews-headline').text
        news2.append(second_new_text)
    except Exception as e:
        print(f"Error extracting second news for {stock}: {e}")
        news2.append('NaN')
        
    try:
        third_new_text = all_news[2].find_element(By.CLASS_NAME, 'LatestNews-headline').text
        news3.append(third_new_text)
    except Exception as e:
        print(f"Error extracting third news for {stock}: {e}")
        news3.append('NaN')
        
    try:
        fourth_new_text = all_news[3].find_element(By.CLASS_NAME, 'LatestNews-headline').text
        news4.append(fourth_new_text)
    except Exception as e:
        print(f"Error extracting fourth news for {stock}: {e}")
        news4.append('NaN')
        
    try:
        fifth_new_text = all_news[4].find_element(By.CLASS_NAME, 'LatestNews-headline').text
        news5.append(fifth_new_text)
    except Exception as e:
        print(f"Error extracting fifth news for {stock}: {e}")
        news5.append('NaN')
        
driver.quit()
```

```python
# Create a DataFrame with the collected data
df = pd.DataFrame({
    'News1': news1,
    'News2': news2,
    'News3': news3,
    'News4': news4,
    'News5': news5
})

# Save the DataFrame to a CSV file
df.to_csv('nasdaq_stocks_news.csv', index=False)
```
